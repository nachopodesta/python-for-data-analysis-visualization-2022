#### Line plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
```

```python
tfl = pd.read_csv('tfl-daily-cycle-hires.csv')
```

```python
tfl.drop(columns='Unnamed: 2', inplace=True)
```

```python
tfl['Day'] = pd.to_datetime(tfl['Day'],format='%d/%m/%Y')
```

```python
tfl.dtypes
```

```python
tfl['Day'] = tfl['Day'].astype('datetime64')
```

```python
tfl.sort_values('Day', inplace=True)
```

```python
tfl['Year'] = tfl['Day'].dt.strftime('%Y')
```

```python
tfl_y = tfl[['Year', 'Number of Bicycle Hires']]
```

```python
tfl_y
```

```python
tfl_y = tfl_y.groupby(by='Year').sum().reset_index()
```

```python
tfl_y
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(tfl_y['Year'],tfl_y['Number of Bicycle Hires'])

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(tfl_y['Year'],tfl_y['Number of Bicycle Hires'],
         tfl_y['Number of Bicycle Hires']*2)

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(tfl_y['Year'],tfl_y['Number of Bicycle Hires'], color='r', alpha=0.2)

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])

ax1.plot('Year','Number of Bicycle Hires', data=tfl_y)

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])

ax1.plot('Year','Number of Bicycle Hires', data=tfl_y, marker='1',markersize=20, linestyle=(0,(1,10)))

plt.show()
```

### 89. Scatter Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('players_21.csv')
```

```python
players.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
players.head()
```

```python
players_100 = players.head(100)
```

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.set_title('Overall ability vs. Wages')
ax1.set_xlabel('Overal Ability')
ax1.set_ylabel('Wages')

ax1.scatter(x=players_100['overall'], y = players_100['wage_eur'], s=players_100['value_eur']/150000)
```

### 90. Histograms

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('Number of players')
ax1.set_ylabel('Overal score (bins)')
plt.xticks([0,25,50,75,100], [0,25,50,75,100])
ax1.hist(players['overall'], bins=[0,25,50,75,100], edgecolor='black')
```

### 91. Box Plots and Violin Plots

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.boxplot(players['overall'], labels=['overall score'])
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.boxplot(players['overall'], labels=['overall score'], showfliers=False, vert=False)
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.boxplot(players['overall'], labels=['overall score'], showfliers=False, vert=False)
plt.show()
```

```python
inter = players[players['club_name'] == 'Inter']['overall']
```

```python
dortmund = players[players['club_name'] == 'Borussia Dortmund']['overall']
```

```python
data = [inter, dortmund]
```

```python
data
```

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.boxplot(data, labels=['Inter', 'Dortmund'])
plt.show()
```

#### Violin PLots

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])

ax1.violinplot(data)
ax1.boxplot(data)
plt.xticks([1,2], ['Inter', 'Dortmund'])

plt.show()
```

### 92. Style and Presentation

```python
import numpy as np
```

```python
x = np.arange(0,10,0.1)
```

```python
x.shape
```

```python
y1 = x
y2 = x**2
y3 = x**3
```

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_title('Chart Title', fontsize=40)
ax1.set_xlabel('x label', fontsize=20, fontname = 'Arial')
ax1.set_ylabel('y label', fontsize=20, fontname = 'Arial')

plt.xticks([0,2,4,6,8,10], ['zero', 'two', 'four', 'six', 'eight','ten'])

plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

plt.annotate('green line', (2,200))
plt.arrow(3,190, dx=2, dy=-60)

plt.show()
```
